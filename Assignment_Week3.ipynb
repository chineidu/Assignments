#### Exercise
Make a plot using the 'Pokemon' dataset and answer the following questions:
* The frequency of Pokemon by type
* The frequency of Pokemon by HP stat total
* The frequency of Pokemon by weight

```python
# Load the data
pkn_data = pd.read_csv('Files/pokemon.csv')

# The first 5 observations of the data
pkn_data.head()
```

```python
# Information about the data
pkn_data.info()
```

<br>

**Qs 1. The frequency of Pokemon by type**

```python
pkn_data['Type 1'].value_counts().plot.bar()
plt.xlabel("Types of Pokemons", fontsize=15)
plt.ylabel("Frequency", fontsize=15)
plt.title("The Frequency of Pokemon by Type", fontsize=17)

plt.show()
```

```python
(pkn_data['Type 1'].value_counts() / len(pkn_data)).plot.bar()
plt.xlabel("Types of Pokemons", fontsize=15)
plt.ylabel("Percentage of Frequency", fontsize=15)
plt.title("The Frequency of Pokemon by Type", fontsize=17)

plt.show()
```

<br>

**Qs2. The frequency of Pokemon by HP stat total**

```python
pkn_data['HP'].value_counts()[:10].sort_index().plot.bar()
plt.xlabel("HP", fontsize=15)
plt.ylabel("Frequency", fontsize=15)
plt.title("The Frequency of Pokemon by HP stat total", fontsize=17)

plt.show()
```

```python
(pkn_data['HP'].value_counts()[:10].sort_index() / len(pkn_data)).plot.bar()
plt.xlabel("HP", fontsize=15)
plt.ylabel("Percentage of Frequency", fontsize=15)
plt.title("The Frequency of Pokemon by HP stat total", fontsize=17)

plt.show()
```
